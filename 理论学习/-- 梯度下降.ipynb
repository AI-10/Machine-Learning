## 梯度下降法

#### 什么是梯度下降？
- 梯度下降法是一种基于搜索的一阶最优化方法，本身不是一种机器学习方法，该算法通常通过参数迭代的方法求一个损失函数得局部极小值从而求得模型的解。

#### 写出梯度下降参数迭代表达式，并解释？

- 梯度下降的原理？
    - 1.根据梯度（导数）的符号来判断最小值点x在哪
    - 2.让函数值下降（变小）。
    - 3.求出参数（而不是函数值）
- α是什么含义？
    - α在梯度下降算法中被称作为学习率或者步长，意味着我们可以通过α来控制每一步走的距离。
- 为什么要梯度要乘以一个负号？
    - 梯度的方向实际就是函数在此点上升最快的方向，而我们需要朝着下降最快的方向走，自然就是负的梯度的方向，所以此处需要加上负号。

#### 梯度下降具体怎么做？
   - （1）首先对θ赋值，这个值可以是随机的，也可以让θ是一个全零的向量。 
   - （2）迭代改变θ的值，使得J(θ)按梯度下降的方向进行减少。
   - （3）满足条件时终止

- 梯度，一个向量，是函数沿坐标方向偏导数组成的向量。
