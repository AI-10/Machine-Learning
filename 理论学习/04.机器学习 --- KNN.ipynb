### 简介：
- KNN(k-nearest neighbor,K近邻算法)，由Cover和Hart于1968年提出，是基本的分类和回归方法，也是最简单的机器学习方法之一，可用于文本分类、模式识别、图像及空间分类等。
- **核心思想**：
    - 分类：如果一个样例在特征空间中的k个最邻近的样例大多数属于某一个类别，则该样例也属于这个类别。
    - 回归：通过找出一个样本的k个最近邻居，将这些邻居的属性的平均值赋给该样例，就可以得到该样例的属性。

### 模型核心

- KNN没有显式的学习过程，主要靠待测样本周围有限的邻近的样例进行预测，而不是通过训练出的不依赖与训练数据的模型进行预测，KNN的模型实际上对应于对特征空间的划分。
- **当训练数据集、距离度量、k值及决策规则确定后，对于新的输入，他的输出唯一确定。**

#### 模型三要素

1.距离度量：反映样本相似程度


最常见距离度量：设特征空间X是n维实数向量空间 ，xi,xj∈X  则：xi,xj的Lp距离定义为:

   
   - p=1 ：欧氏距离
   - p=2 ：曼哈顿距离
   - p=∞：各个坐标距离的最大值

2.K值选择：k值的选择反映了对近似误差与估计误差之间的权衡，对结果影响重大，一般不大于20，通常用交叉验证法选取最优K值。
   - K值较小，近似误差减小，估计误差增大，模型复杂，易过拟合。
   - K值较大，近似误差增大，估计误差减小，模型简单，易欠拟合。

3.决策规则：
   - 分类：投票法or加权投票法
       - 根据给定的距离度量，在训练集中找出与待测样例x最邻近的k个点，涵盖这k个点的邻域记作Nk(x),则有待测样例的分类y(I(.)为指示)：
   
   - 回归：平均法or加权平均法
