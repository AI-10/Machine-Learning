#### 机器学习简介

- 定义：机器学习研究如何通过计算手段，利用经验（机器学习中经验通常以数据的形式存在）来改善系统自身的性能。是一门多领域交叉学科，涉及概率论、统计学、逼近论、凸分析、算法复杂度理论等多门学科。

#### 数据相关术语

- 数据集：示例or记录的集合 = 样本
    - D = {x1,x2,...,xm}
- 训练集：训练数据的集合
- 测试集：测试数据的集合
- 样例：拥有标记信息的示例or记录
- 属性or特征
- 属性值：属性的取值
- 属性空间or样本空间or输入空间:X
    - 属性空间即属性张成空间，每一个样本都可以在此属性张成空间中找到自己的位置
    - 例如：我们用肺活量、1000米测速、跳远三个属性作为描述同学体能的指标，把这三个属性作为三条坐标轴，则这三个指标张成一个用来描述同学体能的三维空间，每一位同学都可以在这个空间中找到自己的坐标位置。
- 标记空间or输出空间：Y
    - 示例标记的集合
    - 例如：根据身高体重判断是否肥胖，则是否肥胖这一属性就是标记，而这所有标记的集合就是标记空间or输出空间。
- 数据维数：属性的数目
- 特征向量：由于每一个张成空间中的点对应一个坐标向量，因此我们也把一个示例称为一个特征向量。

#### 假设空间与版本空间

- 归纳与演绎
    - 归纳（induction：从特殊到一般的“泛化”过程,即从具体的事实归结出一-般性规律)与演绎(deduction：从一般到特殊的“特化”过程,即从基础原理推演出具体状况)是科学推理的两大基本手段，也是重要的学习手段，而现在主流的机器学习方法“从样例中学习”显然是一个归纳的过程，因此所得结果的泛化能力尤为重要，而为了得到良好的泛化能力，我们需要对所有假设进行分析。
    - 例如：第一只瓜是好瓜，即“青绿，蜷缩，清响”的瓜是好瓜，以后碰见一样属性的瓜就是好瓜（机械学习），但我们的目标是泛化，即通过对训练集中的数据进行学习后对没有碰见过的瓜进行判断，例如判断“浅白，蜷缩，浊响”是什么瓜？
    
- 假设空间与版本空间
    - 假设空间：我们可以把学习过程看作一个在所有假设(bypothesis)组成的空间中进行搜索的过程，搜索目标是找到与训练集“匹配”(fit)的假设，即能够将训练集中的瓜判断正确的假设,假设的表示一旦确定, 假设空间及其规模大小就确定了(N1 * N2 * ...* Nm + 1)
    - 版本空间：需注意的是，现实问题中我们常面临很大的假设空间，但学习过程是基于有限样本训练集进行的，因此，可能有多个假设与训练集-致,即存在着一个与训练集一致的“假设集合"，我们称之为“版本空间”(version space)。
