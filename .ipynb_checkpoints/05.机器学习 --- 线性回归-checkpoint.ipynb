### 简介
- **线性回归(Linear Regression)是利用构建线性模型来研究自变量与因变量之间的回归关系，属于参数监督模型，其参数的估计通常是通过最小化平方损失函数来实现的，又叫最小二乘估计。**
    - 线性的含义：因变量（结果变量）可以表示为自变量（属性或特征）的线性函数，即一阶偏导数为常数。
    - 回归的含义：研究一个随机变量Y（为连续值）与另一个或一组变量的相互依存关系的分析方法。（道尔顿）

- **回归方程与回归模型的区别与联系**
    - 区别：
        - 线性回归模型刻画的是总体自变量与因变量之间客观存在的线性回归关系（当这种客观关系存在时），而回归方程是我们在假设空间中根据数据所学到的方程，其本质上是刻画了我们拥有的数据（样本）中的客观关系。
        - 线性回归模型中的Ｙ是一个随机变量，所以正确的表达模型应该在等式右边加上误差项ε来表示Y的变异，而线性回归方程中的Y-hat是一个定数，用来估计线性回归模型中的Y的期望。
    - 联系：
        - 线性回归方程的参数可以作为线性回归模型的估计，但一般要经过假设检验，或者通过模型评价来判断。

其实在机器学习中我们直接用线性回归方程代替线性回归模型，只要注意以我们进行预测时的到的并不是Y的真实值，而是Y真实值的期望估计。

### 策略

- 线性模型的假设空间为

- 而要学习到假设空间中的最优模型，就需要定义线性回归模型的损失函数，即度量模型一次预测的好坏，在线性回归模型中，损失函数定义为平方损失函数（L(X,f(X))=(Y-f(x)）^2，于是我们就得到线性回归模型的风险函数（损失函数度量一次预测的好坏，风险函数度量平均意义下模型预测的好坏）

而线性回归的学习策略就是最小化风险函数，这种求解策略也叫作最小二乘法，其“二乘”指的是用平方来度量观测点与估计点的距离（欧式距离），“最小”指的是参数值要保证各个观测点与估计点的距离的平方和达到最小。

- 为什么要用平方损失函数？
    - 从直观的欧式距离角度出发：平方损失函数是预测值与真实值之间的欧式距离，直观反映了预测数据与真实样本之间的差距，而我们要求预测值尽可能接近真实值，所以在通过最小化平均距离作为我们的求解策略。
    - 从严格的数理逻辑角度求解：实际线性回归模型要求ε-N(μ，σ^2)，那么我们就可以求的参数ε的极大似然估计值，即是说，在某个N(u,σ^2)下使得服从正态分布的ε取得现有样本εi的概率最大，最后求得的结果正好与前面构建的最小化平均损失函数等价。


- 高斯分布是由最后的结果倒推来的，最小二乘的前提误差服从正态分布，如不服从，则用其他损失函数。取绝对值是拉普拉斯分布。虽然有时不服从正态分布，但假设是正态分布效果也不错。

知道了模型与策略，我们下一步就是要进行模型的求解，而求解模型就要了解算法

### 线性回归---正则化

- 线性回归两种正则化
    - L1正则化，lasso回归，按定数衰减，系数可为0
    - L2正则化，ridge回归，按比例衰减

- 复杂度处罚因子的两个作用
    - 不可逆--可逆
    - 大系数--小系数

- 正则化解释
    - 正规方程角度：可逆化（正定化）
    - 梯度下降角度：缩减系数
    - 损失函数角度：函数加值
