```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import scipy as sp
from sklearn.datasets import load_iris
import seaborn as sns
```

```python
data = load_iris().data
```

```python
a = data[:,0]
b = data[:,1]
c = data[:,2]
d = data[:,3]
```

```python
sns.kdeplot(a,color = "black")
sns.kdeplot(b)
sns.kdeplot(c,color = "green")
sns.kdeplot(d,color = "red")
```

```python
print(stats.normaltest(a).pvalue<0.05)
print(stats.normaltest(b).pvalue<0.05)
print(stats.normaltest(c).pvalue<0.05)
print(stats.normaltest(d).pvalue<0.05)
```
