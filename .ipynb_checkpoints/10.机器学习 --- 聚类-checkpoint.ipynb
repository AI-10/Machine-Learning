## 一、简介

#### 1.概念
聚类分析就是将分类对象分成若干类，相似的作为一类，不相似的作为一类，常常用来探索“自然的”分类，显示规律本身固有的类别属性。聚类与分类的不同在于，聚类所要求划分的类是未知的。聚类分析所使用方法的不同，常常会得到不同的结论。不同研究者对于同一组数据进行聚类分析，所得到的聚类数未必一致。

#### 2.解释
- 统计学的角度：聚类分析是通过数据建模简化数据的一种方法。
    - 传统统计聚类分析方法包括系统聚类法、动态聚类法、有序样品聚类、分解法、加入法等。
    - 采用k-均值、k-中心点等算法的聚类分析工具已被加入到许多著名的统计分析软件包中，如SPSS、SAS等。
- 机器学习角度：簇相当于隐藏模式。
    - 聚类是搜索簇的无监督学习过程。与分类不同，无监督学习不依赖预先定义的类或带类标记的训练实例，需要由聚类学习算法自动确定标记。
    - 聚类是观察式学习，而不是示例式的学习。

## 二、相似性度量

- 聚类就是将相似的样本或变量聚在一起，那么样本或变量之间的相似性该怎么度量？常见的度量方法有。
    - 距离：度量样本间的相似性，距离越小越相似
        - 间隔变量：闵可夫斯基距离、马氏距离等
        - 属性变量：匹配程度
    - 相似系数：度量变量之间的相似性，系数越大越相似。

## 三、常见聚类方法
理论上最优的聚类方法似乎是寻求一切可能的聚类，然后依据某种准则，选择最优的聚类方法，但面对大量数据时，这种方法是不可行的（NP完全问题），所以需要寻求一种切实可行且效果较好的聚类方法，此种聚类可能是好的，但未必是最好的，接下来我们介绍两种比较常见的聚类方法。
- 层次聚类法
- 动态聚类法 - k-means聚类
- 密度聚类法 - DBSCAN

### 1.层次聚类法
- 基本思想：层次聚类法又称系统聚类法，是通过一系列相继的合并和分割进行聚类的（接下来主要讲解合并的方法），适用于样本量不是很大的情况。
- 聚类流程：
    - 首先规定样品之间的距离和类与类之间的距离
    - 开始将n个样品各自作为一类
    - 然后将距离最近的两类合并成一个新类
    - 计算新类与其它类的距离(注:未合并的类之间的距离不用重新计算)
    - 重复进行两个最近类的合并，每次减少一类，直至所有的样品合并为一类。
- 分类
    - Q型聚类：针对样本聚类（如：对人群中的人按体型指标进行聚类）
    - R型聚类：针对变量聚类（如：对人体体型指标进行聚类，纵向-身高，肢长等，横向-体重，腰围，腹围等）
- **重点**：可以看出对于层次聚类，最主要的是根据具体情况规定样本间的距离及类之间的距离，而根据类之间的度量方式不同，可以得出不同的系统聚类方法

### 2、动态聚类法 - K-means法

- 基本思想：动态聚类法的基本思想是选择一批凝聚点或者初始的分类，让样本按某种原则向凝聚点聚集，对凝聚点不断地进行修改和迭代，直至分类比较理想或者迭代稳定为止，类的个数一般需要提前指定。动态聚类法有许多的方法，接下来讲解比较流行的一种动态聚类方法 - K-means法(1967.MacQueen) 

- 注意事项：
    - 动态聚类与层次不同的一点是动态聚类对于前一次聚类后的样本可以进行迭代修改，而层次聚类一旦确定类，就无法修改，直到聚类结束。
    - 动态聚类只针对样本，不能用于变量聚类。
    - 初始点及K值需要指定，初始聚类点一般随机抽选（或随机分割），可以要求初始凝聚点具有一定的距离。

- K-means聚类流程
    - 选择k个样品作为初始凝聚点（或者将所有样品分成k个初始类，然后将这k个类的重心(均值)作为初始凝聚点）
    - 对所有的样品逐个归类，将每个样品归入离它最近的那个类(通常采用欧氏距离)，直至所有样品都归了类，然后将该类的凝聚点更新为这一类目前的重心。
    - 重复步骤(2)，直至所有的样品都不能再分配为止。
- 可以看bili上的动画展示https://www.bilibili.com/video/av55177945/?spm_id_from=trigger_reload

- 初始聚集点的选择
    - **初始聚集点对结果有较大的影响，可以多试几个初始点再做判断，若不同聚集点聚类效果完全不同，则可能表明没有自然的类的形成**
    - 初始点选择好一般会有一个好的聚类效果，故K均值法可以用于对系统聚类的改进，如：以系统聚类的类平均法所得聚类的重心作为初始凝聚点，然后进行K均值聚类，不过最后的效果需要结合实际情况

### 3、密度聚类法 - DBSCAN
- 基于密度的聚类假设聚类结构能通过样本分布的紧密程度来确定，通常情况下，密度聚类算法从样本密度的角度考察样本之间的可连接性，并基于可连接样本不断扩展聚类簇获得最终结果，其中DBSCAN是一种著名的密度聚类算法，它基于一组“邻域”参数（ε、MinPts）来刻画样本的紧密程度。

- 给定数据集D={x(1),x(2),...,x(m)}，定义一下几个概念：
    - ε-邻域：（Eps）：对x(j)∈D，其ε-邻域包含D中与x(j)的距离不大于ε的所有样本。
    - MinPts ： ε-邻域内样本个数最小值。
    - 核心对象：若x(j)的ε-邻域至少包含MinPts个样本，|Nε(x(j))|≥MinPts，则x(j)为一个核心对象。
    - 密度直达：若x(j)位于x(i)的ε-邻域中，且x(i)是核心对象，则称x(j)由x(i)密度直达（密度直达关系通常不满足对称性，除非x(j)也是核心对象）
    - 密度可达：对x(i)与x(j)，若存在样本序列p1,p2,...,pn，其中p1=x(i)，pn=x(j)，p1,p2,...,pn-1均为核心对象且pi+1从pi密度直达，则称x(j)由x(i)密度可达。密度可达关系满足直递性，但不满足对称性。
    - 密度相连：对x(i)与x(j)，若存在x(k)使得x(i)与x(j)均由x(k)密度可达，则称x(i)与x(j)密度相连。密度相连关系满足对称性。
    
如图：规定参数（ε，3）则有x2、x1密度直达，x3、x1密度可达，x3、x4密度相连

- DBSCAN算法原理
    - 基于以上概念，DBSCAN将簇定义为：由密度可达关系导出的最大的密度相连样本集合C，簇C满足以下两个性质：
        - 连接性
        - 最大性
- DBSCAN算法流程
    - DBSCAN算法先找出给定邻域参数下的所有核心对象，然后任选数据集中的一个核心对象作为种子，由此出发寻找合并核心对象密度可达的对象，直到所有核心对象均被访问过为止。
    
    - DBSCAN的簇中可以包含一个核心对象：如果只有一个核心对象，则其他非核心对象都落在核心对象的ε-邻域内；如果有多个核心对象，则任意一个核心对象的ε-邻域内至少有一个其他核心对象，否则这两个核心对象无法密度可达；包含过少对象的簇可以被认为是噪音。

## 四、性能度量
- 聚类没有统一的度量方法，因为数据本身的内部结构是很难观察的，所以任何一种评价指标都具有一定的风险，但考虑到聚类的任务的目的，容易想到就是”物以类聚“，即达到”簇内相似度高“，”簇间相似度低“的性能效果，所以，我们通常实际经验及结合多个指标进行判断，常见的性能度量指标有如下两类：
    - 外部指标：将聚类结果与某个“参考模型”进行比较，如JC、FMI、RI等
    - 内部指标：直接考察聚类结果而不用任何参考模型，如DBI、DI、SSE、SC系数、CH系数等

- **对于层次聚类和K-means聚类法，性能度量除了评价聚类效果之外，还可以选择合适的K值（肘部法）**

## 优缺点总结

**1.层次聚类**
- 优点：
    - 距离和规则的相似度容易定义，限制少
    - 不需要预先制定聚类数
    - 可以发现类的层次关系
    - 可以聚类成其它形状
- 缺点
    - 计算复杂度太高
    - 奇异值也能产生很大影响
    - 算法很可能聚类成链状

**2.K-means法**
- 优点：
    - 是解决聚类问题的一种经典算法，简单、快速
    - 对处理大数据集，该算法保持可伸缩性和高效性
    - 当簇接近高斯分布时，它的效果较好。
- 缺点：
    - 算法中 K 是事先给定的，但 K 值的选定是难以估计
    - 初始聚类中心的选择对聚类结果有较大的影响，初始值选择的不好，可能无法得到有效的聚类结果
    - 使用L2距离函数，若簇中含有异常点，将导致均值偏离严重（即:对噪声和孤立点数据敏感）
    - 在簇的平均值可被定义的情况下才能使用，可能不适用于某些应用
    - 不适用于发现非凸形状的簇或者大小差别很大的簇。

**3.DBSCAN**
- 优点：
    - 不需要事先给定簇的数目k
    - 对初始点顺序不敏感
    - 适于稠密的非凸数据集，可以发现任意形状的簇
    - 可以在聚类时发现噪音点、对数据集中的异常点不敏感

- 缺点：
   - 对于高维数据效果不好；
   - 不适于数据集中样本密度差异很小的情况；
   - 调参复杂，给定eps选择过大的MinPts会导致核心对象数量减少，使得一些包含对象较少的自然簇被丢弃；
        - 选择过小的MinPts会导致大量对象被标记为核心对象，从而将噪声归入簇；
        - 给定MinPts选择过小的eps会导致大量的对象被误标为噪声，一个自然簇被误拆为多个簇；
        - 选择过大的eps则可能有很多噪声被归入簇，而本应分离的若干自然簇也被合并为一个簇；
   - 数据量很大时算法收敛的时间较长，可对搜索最近邻时建立的KD-tree或Ball-tree进行规模限制。
